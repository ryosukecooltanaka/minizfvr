## Data check
- check saved data

```python
import numpy as np
from matplotlib import pyplot as plt
import h5py
from pathlib import Path
```

```python
basepath = Path('../stim_dev_v01/saving_test')
fish = '251029_f001'
run = '20251029_142000'
```

```python
with h5py.File(basepath / fish /run/ 'stimulus_log.h5') as f:
    t = f['t'][:]
    phi = f['phi_displacement'][:]
    y = f['y_displacement'][:]
    vigor = f['vigor'][:]
    bias = f['bias'][:]

    plt.plot(t, phi)
    plt.plot(t, y)
    plt.plot(t, vigor*10)
    plt.plot(t, bias*10)
    
    plt.show()
    
```

```python
with h5py.File(basepath / fish / run / 'tail_log.h5') as f:
    t2 = f['t'][:]
    tail = f['tail_angle'][:]
    plt.plot(t2, tail)
    plt.show()
    
```

```python
plt.figure(figsize=(15,5))
plt.plot(t2, tail)
#plt.plot(t, y)
plt.plot(t, vigor)
plt.plot(t, bias)
plt.xlim([0,2])
plt.show()
```
