```python
import numpy as np
from matplotlib import pyplot as plt
import h5py
from pathlib import Path
```

```python
basepath = Path('../stim_dev_v01/saving_test')
```

```python
with h5py.File(basepath / '251024153447' / 'stimulus_log.h5') as f:
    t = f['t'][:]
    phi = f['phi_displacement'][:]
    y = f['y_displacement'][:]
    plt.plot(t[t>0], phi[t>0])
    plt.plot(t[t>0], y[t>0])
    print(len(t[t>0]))
    plt.show()
    
```

```python
with h5py.File(basepath / '251024153447' / 'tail_log.h5') as f:
    t2 = f['t'][:]
    tail = f['tail_angle'][:]
    plt.plot(t2[t2>0]-min(t2[t2>0]), tail[t2>0])
    print(len(t2), np.sum(t2>0))
    plt.show()
    
```
